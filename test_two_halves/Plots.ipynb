```python
f1_sc = pd.DataFrame()
f1_tf = pd.DataFrame()
for i in np.unique(predict_result["sample_size"]):
    col_name = "sample"+str(i)
    temp = predict_result[predict_result["sample_size"]==i]["tf_F1_stream"]
    f1_tf = pd.concat([f1_tf.reset_index(drop=True),temp.reset_index(drop=True)],axis=1)
    temp = predict_result[predict_result["sample_size"]==i]["sc_F1_stream"]
    f1_sc = pd.concat([f1_sc.reset_index(drop=True),temp.reset_index(drop=True)],axis=1)
    
test_tf = f1_tf.T
test_sc = f1_sc.T
```
