# Visualize training and validation data

```python
import numpy as np
import matplotlib.pyplot as plt


X_train = np.load('/content/drive/MyDrive/Transfere Learning research/samples/train_data_aug200_sample.npy').astype(np.float32)
Y_train = np.load('/content/drive/MyDrive/Transfere Learning research/samples/train_label_aug200_sample.npy').astype(np.float32)
```

```python
rando_ind = 155
```

```python
fig, axs = plt.subplots(2, 4,figsize=(20,10))
labels = ["a","b","c","d","e","f","g","h"]
count = 0
img_order = [3,0,1,2,4,5,6,7]
for ax in axs.flat:
  image = X_train[rando_ind,:,:,img_order[count]]
  ax.imshow(image)
  ax.set_xlabel("("+labels[count]+")", fontsize = 16) # X label
  count = count + 1
plt.show()
```

```python
fig, axs = plt.subplots(2, 4,figsize=(20,10))
labels = ["a","b","c","d","e","f","g","h"]
count = 0
for ax in axs.flat:
  image = X_train[rando_ind,:,:, count]
  ax.imshow(image)
  ax.set_xlabel("("+labels[count]+")", fontsize = 16) # X label
  count = count + 1
plt.show()
```
